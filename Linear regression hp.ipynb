```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file=pd.read_csv('B:\honeyproduction.csv')
```

```python
file.head(10)
```

```python
file.describe()
```

```python
hp_year=file[['numcol','yieldpercol','totalprod','stocks','priceperlb','prodvalue','year']].groupby('year').sum()
hp_year.head()
```

```python
hp_year.reset_index(level=0,inplace=True)
hp_year.head()
```

```python
plt.figure(figsize=(16,7))
plt.plot(hp_year['year'],hp_year['yieldpercol'])
plt.title('Trend of Honey Yield per Colony',  fontsize=25)
plt.xlabel('year', fontsize=25)
plt.ylabel('Yield per colony',fontsize=25)
```

```python
plt.figure(figsize=(16,7))
plt.plot(hp_year['year'],hp_year['priceperlb'])
plt.title('Price of honey per lb',  fontsize=25)
plt.xlabel('year', fontsize=25)
plt.ylabel('price pe lb',fontsize=25)
```

```python
plt.figure(figsize=(16,7))
plt.plot(hp_year['year'],hp_year['totalprod'])
plt.title('Total production per year',  fontsize=25)
plt.xlabel('year', fontsize=25)
plt.ylabel('Production',fontsize=25)
```

```python
US_state=file[['state','totalprod','yieldpercol']].groupby('state').sum()
US_state.reset_index(level=0, inplace=True)
US_state.sort_values(by='totalprod', ascending=False,inplace=True)
US_state.head()
```

```python
plt.figure(figsize=(20,7))
sns.barplot(x=US_state['state'], y=US_state['totalprod'])
plt.title('Statewise Total Honey Production in USA', fontsize=20)
plt.xlabel('States',fontsize=20)
plt.ylabel('Total Production of Honey in USA', fontsize=20)
```

```python
plt.figure(figsize=(20,7))
sns.barplot(x=US_state['state'], y=US_state['yieldpercol'])
plt.title('Statewise Total Honey Production in USA', fontsize=20)
plt.xlabel('States',fontsize=20)
plt.ylabel('Yield per colony', fontsize=20)
```

```python
US_state_max=file[['state','totalprod']].groupby('state').max()
US_state_max.reset_index(level=0, inplace=True)
US_state_max.columns=['State','Max Prod']
US_state_max.head()
```

```python
US_state_min=file[['state','totalprod']].groupby('state').min()
US_state_min.reset_index(level=0, inplace=True)
US_state_min.columns=['State','Min Prod']
US_state_min.head()
```

```python
st_range=pd.merge(US_state_max,US_state_min,how='inner',on='State')
st_range.head()
```

```python
st_range['Per_Change']=((st_range['Max Prod']-st_range['Min Prod'])/st_range['Max Prod'])*100
st_range.sort_values(by='Per_Change', ascending=False, inplace=True)
st_range.head()
```

```python
plt.figure(figsize=(20,7))
sns.barplot(x='State', y='Per_Change', data=st_range)
plt.title('Statewise production Decline  Trend', fontsize=20)
plt.xlabel('States',fontsize=20)
plt.ylabel('%decline', fontsize=20)
```

```python
honey_corr=file[['numcol','totalprod','year','yieldpercol','stocks','prodvalue']].corr()
honey_corr
```

```python
sns.heatmap(honey_corr,annot=True, vmin=-1, vmax=1, cmap="YlGnBu")
plt.show()
```

```python
sns.lmplot(x='numcol', y='prodvalue', data=file, hue='year')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(file.drop(['totalprod','state'], axis=1),file['totalprod'], test_size=0.20,random_state=8)
```

```python
model=LinearRegression()
model.fit(X_train, y_train)
```

```python
accuracy=model.score(X_test, y_test)
print(accuracy*100, '%')
```
